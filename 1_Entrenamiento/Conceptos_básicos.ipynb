# 02 Operaciones básicas

```python
# Operaciones básicas
num1 = 10
num2 = 5

# Suma
suma = num1 + num2
print('La suma es: ', suma)

# Resta
resta = num1 - num2
print('La resta es: ', resta)

# Multiplicación
multiplicacion = num1 * num2
print('La multiplicación es: ', multiplicacion)

# División
division = num1 / num2
print('La división es: ', division)

# Potenciación
potenciacion = num1 ** num2
print('La potenciación es: ', potenciacion)

# Módulo
modulo = num1 % num2
print('El módulo es: ', modulo)
```

# 05 Estructuras de control

```python
# Condicional if, elif y else
edad = 20
if edad < 18:
    print('Menor de edad')
elif edad == 18:
    print('Tienes 18 años')
else:
    print('Mayor de edad')

# Bucles
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(numero)
print('Fin del bucle')

# Bucle while
contador = 0
while contador < 5:
    print(contador)
    contador += 1
print('Fin del bucle')

# Ejercicio - Determinar numeros positivos, negativos o cero
numero = int(input('Ingrese un numero: '))
if numero > 0:
    print('El numero es positivo')
    for i in range(1, numero, 1):
        print(i)
elif numero < 0:
    print('El numero es negativo')
    while numero < 0:
        print(numero)
        numero += 1
else:
    print('El numero es cero')
print('Fin del programa')
```

# 06 Listas y Tuplas

```python
# Crear lista
mi_lista = [10, 20, 30, 40, 50]

# Acceder a elementos de la lista
print(mi_lista)
print('Elemento en la posición 3: ', mi_lista[2])

# Modificar elementos de la lista
mi_lista[2] = 100
print(mi_lista)

# Agregar un elemento a la lista
mi_lista.append(150)
print('append: ', mi_lista)

# Remover un elemento de la lista
mi_lista.remove(40)
print('remove: ', mi_lista)

# Eliminar el último elemento
mi_lista.pop()
print('pop: ', mi_lista)

# Ordenar la lista
mi_lista.sort()
print('sort: ', mi_lista)



# Crear tupla
mi_tupla = (1, 2, 3, 4, 5)

# Acceder a elementos de la tupla
print('Tupla: ', mi_tupla)
print('Elemento en la posición 3: ', mi_tupla[2])

# La tupla no permite modificar sus valores
# mi_tupla[1] = 20
```

# 07 Diccionarios y Conjuntos

```python
# Crear un diccionario
mi_diccionario = {
    'nombre': 'Holguer',
    'edad': 30,
    'ciudad': 'Pasto',
    'bilingue': False
}

# Acceder a valores del diccionario
print('Diccionario', mi_diccionario)
print('Nombre: ', mi_diccionario['nombre'])
print('Edad: ', mi_diccionario['edad'])

# Modificar elementos
mi_diccionario['edad'] = 15
print('Edad modificada: ', mi_diccionario['edad'])

# Métodos comunes en diccionarios
print('Keys: ', mi_diccionario.keys())
print('Values: ', mi_diccionario.values())
print('Items: ', mi_diccionario.items())

# Agregar una nueva propiedad
mi_diccionario['pais'] = 'Colombia'
print('Pais agregado: ', mi_diccionario)

# Eliminar una propiedad
mi_diccionario.pop('ciudad')
print('Pop: ', mi_diccionario)


# Conjuntos - No permite valores duplicados
# Crear un conjunto
mi_conjunto = {1, 2, 3, 4, 5}
print('Conjunto', mi_conjunto)

# Union
otro_conjunto = {4, 5, 6, 7, 8}
union = mi_conjunto.union(otro_conjunto)
print('Union: ', union)

# Interseccion
interseccion = mi_conjunto.intersection(otro_conjunto)
print('Interseccion: ', interseccion)

# Diferencia
diferencia = mi_conjunto.difference(otro_conjunto)
print('Diferencia: ', diferencia)

# Métodos comunes en conjuntos
# Agregar un elemento
mi_conjunto.add(6)
print('Add: ', mi_conjunto)

# Eliminar un elemento | Causa error al intentar eliminar un elemento que no existe
mi_conjunto.remove(2)
print('Remove: ', mi_conjunto)

# Actualizar un elemento
mi_conjunto.update([7, 8, 9]) # Agrega los valores al final
print('Update: ', mi_conjunto)

# Descartar un elemento | No causa error al intentar eliminar un elemento que no existe
mi_conjunto.discard(9)
print('Discard: ', mi_conjunto)

```

# 08 Funciones

```python
# Declarar función
def saludar():
  print('Hola')

# Llamar función
saludar()

# Declarar función con parametros
def saludar_nombre(nombre):
  print('Hola', nombre)
  print(f'Hola {nombre}')

# Llamar función
saludar_nombre('Holguer')

# Función con retorno de valores
def sumar(a, b):
  return a + b

# Llamar función
resultado = sumar(10, 20)
print('Resultado:', resultado)

# Función anonima o lambda
multiplicar = lambda a, b: a * b
# Llamar función
resultado = multiplicar(10, 20)
print('Resultado:', resultado)
#
```
